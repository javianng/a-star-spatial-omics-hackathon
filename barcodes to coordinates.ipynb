```python
import csv
import gzip
import os
import scipy.io
import os.path
from pathlib import Path
import pandas as pd
```

```python
# Path to your CSV file
csv_file_path = 'tissue_positions.csv'

# Making an empty list
barcode = []

# Open the CSV file in read mode
with open(csv_file_path, 'r') as csv_file:
    # Create a CSV reader object
    csv_reader = csv.reader(csv_file)
    
    # Skipping column header
    next(csv_reader)
    
    # Iterate over each row in the CSV file
    for row in csv_reader:
        # Each row is a list of values
        barcode.append(row)

barcode_coord = list(map(lambda x: [x[0], x[1], x[2], x[3]], barcode))
barcode_wcell = list(filter(lambda x: x[1] == '1', barcode_coord))

# Checking
#barcode_wcell
```

```python
barcode_dict = {}
for dots in barcode_wcell:
    if dots[0] != barcode_dict:
        barcode_dict[dots[0]] = dots[2] + ', ' + dots[3]

# Checking
#barcode_dict
```

```python
## This block will take a while to run ##
cd = Path(os.path.abspath(""))

# define MEX directory
matrix_dir = os.path.join(cd, "CytAssist_FFPE_Protein_Expression_Human_Glioblastoma_filtered_feature_bc_matrix", "filtered_feature_bc_matrix")
# read in MEX format matrix as table
mat_filtered = scipy.io.mmread(os.path.join(matrix_dir,"matrix.mtx.gz"))

# list of transcript ids, e.g. 'ENSG00000187634'
features_path = os.path.join(matrix_dir, "features.tsv.gz")
feature_ids = [row[0]  for row  in csv.reader(gzip.open(features_path, mode="rt"), delimiter="\t")]

# list of gene names, e.g. 'SAMD11'
gene_names = [row[1]  for row  in csv.reader(gzip.open(features_path, mode="rt"), delimiter="\t")]

# list of feature_types, e.g. 'Gene Expression'
feature_types = [row[2]  for row  in csv.reader(gzip.open(features_path, mode="rt"), delimiter="\t")]

# list of barcodes, e.g. 'AAACATACAAAACG-1'
barcodes_path = os.path.join(matrix_dir, "barcodes.tsv.gz")
barcodes = [row[0]  for row  in csv.reader(gzip.open(barcodes_path, mode="rt"), delimiter="\t")]
barcodes = [barcode_dict[i] for i in barcodes]
```

```python
poo = pd.DataFrame.sparse.from_spmatrix(mat_filtered)
poo.columns = barcodes
poo.insert(loc=0, value=feature_ids, column="feature id")
poo.insert(loc=1, value=gene_names, column="gene name")

poo
```
